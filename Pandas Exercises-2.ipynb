```python
df=pd.read_table('wind.data',delim_whitespace=True,names=list(range(1,15)),header=0)
df.head()
```

### Step 3. Assign it to a variable called wine

```python
wine=df
wine.head()
```

### Step 4. Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns

```python
wine.shape
```

```python
wine=wine.drop(columns=[1,4,7,9,11,13,14],axis=1)
wine.head()
```

### Step 5. Assign the columns as below:

The attributes are:  
1) alcohol  
2) malic_acid  
3) alcalinity_of_ash  
4) magnesium  
5) flavanoids  
6) proanthocyanins  
7) hue 

```python
wine=wine.rename(columns={2:'alcohol',3:'malic_acid',5:'alcalinity_of_ash',6:'magnesium',8:'flavanoids',10:'proanthocyanins',12:'hue'})
wine.head()
```

### Step 6. Set the values of the first 3 rows from alcohol as NaN

```python
wine.iloc[0:3,0]=np.nan
```

```python
wine.head()
```

```python
wine.loc[0:3,'alcohol']=np.nan
```

```python
wine.head()
```

### Step 7. Now set the value of the rows 3 and 4 of magnesium as NaN

```python
wine.iloc[2:4,3]=np.nan
```

```python
wine.head(6)
```

```python
wine.loc[2:4,'magnesium']=np.nan
```

```python
wine.head(6)
```

### Step 8. Fill the value of NaN with the number 10 in alcohol and 100 in magnesium

```python
wine.alcohol=wine.alcohol.fillna(10)
wine.magnesium=wine.magnesium.fillna(100)
```

```python
wine.head(6)
```
